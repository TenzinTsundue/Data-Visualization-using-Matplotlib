## Scatter Plot

Scatter plot helps in visualizing 2 numeric variables. It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. It also helps in detecting outliers in the plot.

<strong>When to use: </strong> It is used in Machine learning concepts like regression, where x and y are continous variables. It is also used in clustering scatters or outliers detction.

<tt>plt.scatter()</tt>

```python
# Importing Libraries and Data

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries = pd.read_csv('countries.csv')
```

```python
countries.head()
```

```python
data_2007=countries[countries['year']==2007]
```

#### Scatter Plot

```python
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
data_2007['gdpPerCapita'].corr(data_2007['lifeExpectancy'])
```

```python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
```

#### Scater Plot

3D scatterplot helps in visualizing 3 numerical variables in a three dimensional plot

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'],countries['gdpPerCapita'],countries['lifeExpectancy'], s=30)
plt.show()
```
