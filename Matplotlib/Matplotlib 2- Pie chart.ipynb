# Pie chart

It is a circular plot which is divided into slices to illustrate numerical proportion of parts out of a whole

<tt>plt.pie()</tt>

```python
# improrting Libraries and Data

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov = pd.read_csv('types_movies.csv')
```

```python
t_mov
```

```python
# Pie chart
# Pie chart shows distribution of data based on proportion of pie occupied.

plt.pie(t_mov['Percentage'],
       labels = t_mov['Sector'])
plt.axis('equal')
plt.show()
```

##### Pie chart componets
- <strong>wedges:</strong> A list of Patch objects representing each wedge
- <strong>texts:</strong> List of Text objects representing the labels
- <strong>autotexts:</strong> List of Text objects for the numeric values - this is only available if the autopct value for the pie chart is not None

```python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'],
                                   labels = t_mov['Sector'],
                                  autopct = '%.2f')
plt.axis('equal')

print('Wedge: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

#### Pie chart customizations

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered', 'chocolate']

plt.pie(t_mov['Percentage'],
       labels = t_mov['Sector'],
       colors = colors,
       autopct = '%.2f')

plt.axis('equal')

plt.show()
```

#### The explode property

To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart


```python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```
